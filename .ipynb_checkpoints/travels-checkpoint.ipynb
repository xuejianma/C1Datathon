```python
mypath = './datasets_full/Rio/'
file = 'brazil_international_arrivals.csv'

df3 = pd.read_csv(mypath+file)
df3.value.astype('str')
df3['value'] = df3['value'].str.replace(',','')
df3['value'] = df3['value'].astype(float)
# plt.plot(df3.groupby('year').sum(),'.')
df3['ym'] = df3['year']+(df3['month']-1)/12
plt.plot(df3.groupby('ym').sum()['value'],'-')
plt.title('Rio Visits')
```

```python
import os
from datetime import timedelta
from datetime import datetime
def conv(value_):
    try :
        return float(value_)
    except ValueError:
        try:
            assert isinstance(value_, str)
#             print(value_,float(value_.replace(",", "")))
            value_ = float(value_.replace(",", ""))
            return value_
        except:
            print(value_, type(value_))
            return np.nan
            
RIO_DATA_PATH = './datasets_full/Rio/'
brazil_arr = pd.read_csv(os.path.join(RIO_DATA_PATH, "brazil_international_arrivals.csv"))
brazil_arr["date"] = brazil_arr.apply(lambda d: datetime.strptime("{}{}01".format(d["year"], d["month"]), "%Y%m%d"), axis=1)
brazil_arr["value"] = brazil_arr["value"].apply(conv)
brazil_arr.drop(["year", "month"], axis=1, inplace=True)
# brazil_arr.dropna(subset=["value"], inplace=True)
brazil_arr
```

```python
def fun(x):
    x = x+1
y = 2
fun(y)
y
```

```python
agg_series = brazil_arr.groupby("date")["value"].sum().to_frame()
agg_series
# agg_series.shape  # (330, 2)  instead of 358
date_ =pd.DataFrame(index=pd.date_range(agg_series.index.min(), agg_series.index.max(), freq="1BM")).reset_index()
date_ = pd.DataFrame(index=date_["index"].apply(lambda x : x.replace(day=1)))
agg_series = date_.join(agg_series)
agg_series.index.name = "date"
agg_series.head(20)
```

```python
mypath = './datasets_full/London/'
file = 'UK_international-visits.csv'
df = pd.read_csv(mypath+file)
df = df.replace('2019P','2019')
df = df.replace('2019',2019)
df = df.replace('2018',2018)

# df[df['year']==2011].groupby('area').sum()['sample'].hist()
l1 = df[df['year']==2011].groupby('area').apply(lambda data: sum(data['visits']*data['sample']))
l2 = df[df['year']==2012].groupby('area').apply(lambda data: sum(data['visits']*data['sample']))
l3 = df[df['year']==2013].groupby('area').apply(lambda data: sum(data['visits']*data['sample']))

plt.plot(l1,'.',label=2011)
plt.plot(l2,'.',label=2012)
plt.plot(l3,'.',label=2013)
plt.xticks(rotation=90)
plt.legend()

# df['area'].hist()
# plt.xticks(rotation=90)

```
