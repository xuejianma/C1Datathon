```python
mypath = './datasets_full/London/'
file = 'UK_international-visits.csv'
df = pd.read_csv(mypath+file)
df = df.replace('2019P','2019')
df = df.replace('2019',2019)
df = df.replace('2018',2018)

# df[df['year']==2011].groupby('area').sum()['sample'].hist()
l1 = df[df['year']==2011].groupby('area').apply(lambda data: sum(data['visits']*data['sample']))
l2 = df[df['year']==2012].groupby('area').apply(lambda data: sum(data['visits']*data['sample']))
l3 = df[df['year']==2013].groupby('area').apply(lambda data: sum(data['visits']*data['sample']))

plt.plot(l1,'.',label=2011)
plt.plot(l2,'.',label=2012)
plt.plot(l3,'.',label=2013)
plt.xticks(rotation=90)
plt.legend()

# df['area'].hist()
# plt.xticks(rotation=90)

```
